# Atividade 2 -  Polarização por reflexão em um dielétrico

```python
import numpy as np
import pandas as pd
import exp_physics as ep
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.optimize as scp
```

```python
I = lambda theta, I0, eta, xi : I0*(1 - eta * np.sin(2*theta) + xi * np.cos(2*theta))
```

```python
df25 = pd.read_csv('dados/dados25.csv')

df45 = pd.read_csv('dados/dados45.csv')

df65 = pd.read_csv('dados/dados65.csv')
```

```python
df25['theta'] = df25['theta'].apply(lambda x: x*np.pi/180)

df45['theta'] = df45['theta'].apply(lambda x: x*np.pi/180)

df65['theta'] = df65['theta'].apply(lambda x: x*np.pi/180)
```

```python
fit25, cov25 = scp.curve_fit(f = I, xdata = df25['theta'], ydata = df25['I'])
```

```python
fit25
```

```python
ep.residual_plot(df25['theta'], df25['I'], I, fit25, 'Gráfico para incidência de 25°', residual_title='Resíduos', y_title='Intensidade (V)', x_title='Ângulo (rad)')
```
